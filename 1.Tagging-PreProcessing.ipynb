```
#===========Parameters==============
train_size = 5000
test_size = 15000
#===========Parameters==============


#===========Files==============
csv_file = 'drive/My Drive/Colab Notebooks/dat/output-maintenance-item.csv'
#expert_file = 'drive/My Drive/Colab Notebooks/dat/expertTestFile.txt'
train_file = 'drive/My Drive/Colab Notebooks/dat/train.txt'
validation_file = 'drive/My Drive/Colab Notebooks/dat/validate.txt'
test_file = 'drive/My Drive/Colab Notebooks/dat/Test.txt'
#===========Files==============

import csv
import re

shorttext = []
#the list of work orders which were manually tagged so they must be excluded from the training and test data
exclude = [674,5660,6146,8384,12522,14346,14447,15624,20339,21917,22202,24693,
          24778,25063,25101,28310,30282,32115,40310,40331,40342,41402,43895,44193,
          44523,48962,49170,51827]
count = 0

#read file in and save into list
#'''
with open(csv_file, newline='') as csvfile:
    for num, line in enumerate(csvfile.readlines()):
        if count in exclude:
          count = count+1
          continue
        array = line.split(',')
        shorttext.append([num,array[1]])
        count = count+1
        if (count>=train_size + test_size):
            break

#remove header "shorttext"
shorttext.pop(0)


'''
with open(expert_file, 'r') as e:
  for line in e:
    shorttext.append(line)
    print(line)
'''


count =0
with open(train_file, 'w') as t,open(validation_file, 'w') as v,open(test_file, 'w') as test:
    for i,instance in enumerate(shorttext):
        text = instance[1]
        entry = ""
        for word in text.split():
            
            #if an item word
            if ("~" in word):
                items = word.split("~")
                for num, item in enumerate(items):
                    if(num==0):
                        entry = entry + item+"~B-Item "
                    elif (item != ""):
                        entry = entry + item+"~I-Item "

            #if an action word
            elif ("=" in word):
                acts = word.split("=")
                for num, act in enumerate(acts):
                    if(num==0):
                        entry = entry + act+"~B-Act "
                    elif (act != ""):
                        entry = entry + act+"~I-Act "
                
            #if a status word
            elif ("#" in word):
                stats = word.split("#")
                for num, stat in enumerate(stats):
                    if(num==0):
                        entry = entry + stat+"~B-Stat "
                    elif (stat != ""):
                        entry = entry + stat+"~I-Stat "
            else:
                entry = entry + word+"~O "
                prev = "o"
        entry=entry+"\n"
        #save to training file
        if count<(train_size*8/10):
            t.write(entry)
        #save to validation file
        elif count<(train_size):
            v.write(entry)
        #save to test file
        elif count< train_size + test_size:
            test.write(str(i) + ":" + entry)
        count = count+1
        prev = "o"
```
