```

import nltk
nltk.download('punkt')

from nltk.tokenize import sent_tokenize, word_tokenize 
#import warnings 
#warnings.filterwarnings(action = 'ignore')
import gensim 
from gensim.models import Word2Vec
  

inputfile = 'drive/My Drive/Colab Notebooks/dat/neo4jTriples-detailed.csv'
item= []
loc = []
text= []
funcloc=[]
pred = []
size = 5000
test_size = 5000

with open(inputfile, 'r') as f:
    for line in f:
        entries = line.split(',')
        item.append(entries[1])
        loc.append(entries[6])
        funcloc.append(entries[5][10:])
        text.append(entries[1])
        #text.append(entries[8].replace("\n", " ")+entries[5][10:])
        
FL=set(funcloc)
FL.remove('')
print(text[6])


data = [] 
for i in range(size):
    t = text[i]
    temp = [] 
      
    for word in word_tokenize(t): 
        temp.append(word.lower()) 
    data.append(temp)

# Create model 
model = gensim.models.Word2Vec(data, min_count = 1,  
                              size = 100, window = 5)

#print('Test:', model.similarity('mast', '050-081'))

for k in range(test_size):
    entry = item[size+k]
    max = 0
    maxloc = ''
    for location in FL:
        try:
            val=model.similarity(entry, location)
            if val>max:
                max = val
                maxloc = location
        except KeyError:
            continue
    pred.append(maxloc)
    
acc =0
for i in range(test_size):
    if(pred[i] !=funcloc[size+i] and pred[i] !=''):
      print(item[size+i], ' | ', pred[i], ' | ',funcloc[size+i])
    elif pred[i] == funcloc[size+i]:
      acc = acc+1
      
print('Final Accuracy: ',acc/10000)
```
